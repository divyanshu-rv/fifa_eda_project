```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
fifa = pd.read_csv('fifa.csv')
```

```python
fifa
```

```python
fifa.shape
```

```python
fifa[['Value','Wage']]
```

```python
a = fifa.Value
```

```python
lst=[]
for i in a :
    lst.append(i[1:-1])
lst    
```

```python
fifa['Value'] = np.array(lst)
```

```python
b=fifa.Wage
```

```python
lst_1=[]
for i in b :
    lst_1.append(i[1:-1])
lst_1   
```

```python
fifa['Wage'] = np.array(lst_1)
```

```python
fifa['Release Clause'].fillna('€0M',inplace=True)
```

```python
c = fifa['Release Clause']
```

```python
lst_2 = []
for i in c :
    lst_2.append(i[1:-1])
lst_2    
```

```python
fifa['Release Clause'] = np.array(lst_2)
```

```python
fifa["Value"] = pd.to_numeric(fifa["Value"], downcast="float")
fifa["Wage"] = pd.to_numeric(fifa["Wage"], downcast="float")
fifa["Release Clause"] = pd.to_numeric(fifa["Release Clause"], downcast="float")
```

```python
fifa.Joined.fillna(0,inplace=True)
```

```python
d = fifa.Joined
```

```python
fifa.Joined.replace(to_replace=0,value='####',inplace=True) 
```

```python
lst_3 = []
for i in d :
    lst_3.append(i[-4:])
lst_3    
```

```python
fifa.Joined.replace(to_replace='####',value='2000',inplace=True)            ##replacing null values with dummy 2000
```

```python
fifa['Joined'] = np.array(lst_3)
```

```python
fifa['Contract Valid Until'] = pd.to_datetime(fifa['Contract Valid Until'])
```

```python
x = fifa["Height"][0][0]
```

```python
y = fifa["Height"][0][-1]
```

```python
z=x+'.'+y
z
```

```python
fifa.Height.fillna('00',inplace=True)
```

```python
a = fifa.Height
```

```python
fifa.Height
```

```python
lst_4 = []
for i in a :
    x = i[0]+'.'+i[2:]
    lst_4.append(x)
lst_4    
```

```python
fifa['Height'] = np.array(lst_4)
```

```python
fifa["Height"] = pd.to_numeric(fifa["Height"], downcast="float")
```

```python
d = fifa['Weight']
```

```python
fifa.Weight.fillna('000000',inplace=True)
```

```python
lst_5 = []
for i in d :
    x = i[0:-3]
    lst_5.append(x)
lst_5    
```

```python
fifa['Weight'] = np.array(lst_5)
```

```python
fifa["Weight"] = pd.to_numeric(fifa["Weight"], downcast="float")
```

```python
fifa.Joined.replace(to_replace='####',value='2000',inplace=True)
```

```python
fifa['Joined'] = fifa['Joined'].astype(int)
```
